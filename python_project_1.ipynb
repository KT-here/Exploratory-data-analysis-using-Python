```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dataset = pd.read_excel("Superstore_USA.xlsx")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
# finding missing data
dataset.isnull().sum()
```

```python
#filling missing data with mean value
dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean(),inplace=True)
```

```python
#counting the number of rows in that column with same values
dataset['Order Priority'].value_counts()
```

```python
#checking for all the unique values
dataset['Order Priority'].unique()
```

```python
#replacing the value 
dataset['Order Priority']=dataset['Order Priority'].replace('Critical ','Critical')
```

```python
#plotting the graph
plt.figure(figsize=(6,4))  #gives size to graph
sns.countplot(x="Order Priority",data=dataset)
plt.title("count of Order Priority")
plt.savefig("count of Order Priority")
plt.show()
```

# ship mode

```python
#counting the number of rows in that column with same values
dataset['Ship Mode'].value_counts()
```

```python
x=dataset['Ship Mode'].value_counts().index
y=dataset['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(6,4))
plt.pie(y,labels=x, startangle=45, autopct="%0.2f%%")
plt.legend(loc=2)
plt.savefig("Shipping mode")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Ship Mode",data=dataset, hue="Product Category")
plt.savefig("Shipping mode'(Product)'")
plt.show()
```

# customer segment

```python
#plotting the graph
plt.figure(figsize=(6,4))  #gives size to graph
sns.countplot(x="Customer Segment",data=dataset)
plt.savefig("Customer Segment")
plt.show()
```
